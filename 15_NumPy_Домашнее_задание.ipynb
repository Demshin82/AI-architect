```python
Matrix_8x8 = np.zeros((8,8))      # Создаём нулевую матрицу

Slice1 = Matrix_8x8[::2,1::2]     # Создаём срез чётных строк
Slice2 = Matrix_8x8[1::2,::2]     # Создаём срез не чётных строк
for i in range(0,4):              # Заполняем срезы значениями
  for j in range(0,4):
    Slice1[i,j] = 1
    Slice2[i,j] = 1

print(Matrix_8x8)                 # Выводим исходную матрицу

```

```python
Vector1 = np.arange(0,1.5,0.1)
Vector2 = np.arange(1,7,1)
print(Vector1, Vector2)
```

```python
Reshaped_Vector1 = Vector1.reshape((5,3))                     #преобразуем вектор №1
Reshaped_Vector2 = Vector2.reshape((3,2))                     #преобразуем вектор №2
VectorsMultiplication = Reshaped_Vector1 @ Reshaped_Vector2   #Умножаем вектора через оператор "@"
#print(Reshaped_Vector1)
#print(Reshaped_Vector2)
print(VectorsMultiplication)                                  # Выводим результат перемножения
```

```python
Vector_Int32 = np.arange(1, 36*2 ,2, dtype=np.int32)  # Создаём исходный вектор
Vector_temp = np.zeros(36, dtype=np.int32)            # Создаём временный вектор

j=0                                                   # Инициализируем индексную переменную для "отлова" значений кратным 3
for i in range(0,Vector_Int32.size):                  # Начинаем перебор значений в векторе (массиве)
  if Vector_Int32[i] % 3 == False:                    # Проверяем условие кратности 3
    Vector_temp[j] = Vector_Int32[i]                  # Если условие выполняется, присваиваем значение исходного вектора следующему значению временного вектора
    j=j+1                                             # Инкрементируем переменную индекса временного массива

Slice = Vector_temp[:j:]                              # Дераем срез временного массива "Удаляя" нулевые значения

print('Исходный вектор - ', Vector_Int32)             # Выводим исходный вектор
print('результирующий вектор - ', Slice)              # Выводим результирующий вектор (по факту - срез)

```
