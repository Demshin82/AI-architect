```python
Vector_Int32 = np.arange(1, 36*2 ,2, dtype=np.int32)  #   
Vector_temp = np.zeros(36, dtype=np.int32)            #   

j=0                                                   #     ""   3
for i in range(0,Vector_Int32.size):                  #      ()
  if Vector_Int32[i] % 3 == False:                    #    3
    Vector_temp[j] = Vector_Int32[i]                  #   ,        
    j=j+1                                             #     

Slice = Vector_temp[:j:]                              #     ""  

print('  - ', Vector_Int32)             #   
print('  - ', Slice)              #    (  - )

```
