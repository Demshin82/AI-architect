```python
import pandas as pd
import numpy as np
```

Загружаем ДатаСет TITANIC в объект DataFrame

```python
DS = pd.read_csv('Titanic-Dataset.csv')
```

Подсчитываем кол-во пустых ячеек

```python
DS_empty = pd.isna(DS).sum()       # Помещаем в объекст Series колво пустых ячеек по столбцам
EmptyItemsCount = DS_empty.sum()   # Помещаем в Int64 переменную общего кол-ва пустых строк
```

```python
EmptyItemsCount               # переменная содержащая пустые строки
```

Заполняем пустые ячейки

```python
DS_Filled = DS.fillna(value='NoData')   # Заполняем пустые ячейки объекта DataFrame значением "NoData"
```

```python
DS_Filled
```

Аналитика

```python
DS_unique = DS.nunique()         # Записываем в объект Series кол-во уникальныых записей по каждому столбцу
DS_unique.to_frame()             # Преобразуем объект Series в DataFrame (для удобства отображения)
```

```python
DS_max = DS.max()           # Записываем в объект Series максимальные значения по каждому столбцу
DS_max.to_frame()           # Преобразуем объект Series в DataFrame (для удобства отображения)
```

```python
DS_min = DS.min()             # Записываем в объект Series минимальные значения по каждому столбцу
DS_min.to_frame()             # Преобразуем объект Series в DataFrame (для удобства отображения)
```

Производим сводную аналитику

```python
DS.pivot_table(values='Survived', index=['SibSp'], aggfunc="sum" ) # делаем сводную аналитику - кол-во выживших
                                                                   # в зависимости от кол-ва братьев, сестер или супругов на борту у человека
```

```python
DS.pivot_table(values='Survived', index=['Parch'], aggfunc="sum" ) # делаем сводную аналитику - кол-во выживших
                                                                   # в зависимости от кол-ва родителей или детей,скоторыми путешествовал каждый пассажир
```
